```python
# Import Libraries

# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configuration of the plots we will create

#Read in data

df = pd.read_csv(r'L:\Projects\Movie_Industry\movies.csv')
```

```python
df.head()
```

```python
# Finding if any missing data exists

for col in df.columns:
    print(df[col].isnull().value_counts(), "\n")
```

```python
#Data types of our colums

df.dtypes
```

```python
#Filling null values with zero

df = df.dropna()
```

```python
# Removing non required zero from end of budget and gross column by changing the data type as it is not required

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# changing the data type of released year and storing it in new column

df['correctyear'] = df['released'].astype(str)
```

```python
df
```

```python
# Just getting the year value and storing it in correct year

df['correctyear'] = df['correctyear'].str.split().str[2]
```

```python
df
```

```python
df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
df_numerized = df.copy()

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
